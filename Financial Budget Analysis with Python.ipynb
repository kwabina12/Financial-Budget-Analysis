# What is a Financial Budget?

In short, it is a detailed report on the income and expenditure of the government for a financial year.

You may get the task of analyzing a country’s financial budget every year if you are working as a data analyst in the media and communications field, as the media have to explain the government’s priorities for the complete financial year.

I hope you now have understood what is a financial budget and when you may need to analyze it as a data analyst. Let’s see how we can perform the task of financial budget analysis with Python. I will start this task by importing the necessary Python libraries and a dataset that contains data about the financial budget of India for the year 2021:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r"C:\Users\HP\Desktop\Openlabs\DATASET\Financial Budget DataSet.xlsx")
data.head()
```

Let’s have a look at all the departments that are covered in this budget

```python
print(data)
```

 NaN value in this dataset, let’s remove the NaN values and continue with the task of financial budget analysis with Python

```python
data.dropna()
```

not all the departments that are covered in this dataset are the main departments, as some departments can be covered in the others category. So let’s prepare the data by only selecting the main departments and putting all the other departments in the other category

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
data = data.append(row, ignore_index = True)
print(data)
```

 let’s plot this data to have a look at the priorities of the government for the financial year:

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

 the finance department is getting the most of the share from the total budget of the government. Now let’s plot this data into a donut plot to have a clear view of the distribution of funds among all the departments

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
